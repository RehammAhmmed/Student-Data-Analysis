# Student Data Analysis

In this notebook, we will analyze student performance data and visualize it using bar charts.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Entering student and subject data
data = {
    'Student Name': ['Ahmed', 'Sara', 'Mohamed', 'Fatima', 'Ali'],
    'Subject 1': [80, 70, 90, 85, 78],
    'Subject 2': [75, 80, 70, 65, 88],
    'Subject 3': [60, 75, 85, 90, 70],
    'Difficulty of Subject 1': [0.6, 0.5, 0.7, 0.4, 0.6],
    'Difficulty of Subject 2': [0.7, 0.6, 0.5, 0.8, 0.7],
    'Difficulty of Subject 3': [0.5, 0.4, 0.6, 0.5, 0.8]
}

# Convert data to DataFrame
df = pd.DataFrame(data)
df.set_index('Student Name', inplace=True)
```

# Average grades

1. 
We calculate the average grades for each subject and visualize them using a bar chart.

```python
# Calculate average grades for each subject
grades = df[['Subject 1', 'Subject 2', 'Subject 3']].mean()

# Plot Bar Chart for average grades
plt.figure(figsize=(10, 6))
grades.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Average Grades of Students in Each Subject')
plt.xlabel('Subjects')
plt.ylabel('Average Grade')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

# Pass Rates

We calculate the percentage of students who passed each subject (with a grade above 60) and visualize the pass rates.

```python
# Calculate pass rates (percentage of grades above 60)
pass_rate = df[['Subject 1', 'Subject 2', 'Subject 3']].apply(lambda x: (x > 60).mean() * 100)

# Plot Bar Chart for pass rates
plt.figure(figsize=(10, 6))
pass_rate.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Pass Rate for Each Subject')
plt.xlabel('Subjects')
plt.ylabel('Pass Rate (%)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

# Pass Rates

We calculate the percentage of students who passed each subject (with a grade above 60) and visualize the pass rates.

```python
# Calculate average difficulty levels for each subject
difficulty = df[['Difficulty of Subject 1', 'Difficulty of Subject 2', 'Difficulty of Subject 3']].mean()

# Plot Bar Chart for difficulty levels
plt.figure(figsize=(10, 6))
difficulty.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Difficulty Levels of Each Subject')
plt.xlabel('Subjects')
plt.ylabel('Average Difficulty')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
